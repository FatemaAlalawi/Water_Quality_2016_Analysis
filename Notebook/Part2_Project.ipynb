# Part (2): Changes in Parameters that Affect Water Quality Over Time

#### 1. Safety Standards for Water Quality

- **pH**: 6.5 â€“ 8.5  
- **Temperature**: < 25Â°C  
- **Turbidity**: < 5 NTU (ideally < 1 NTU)  
- **Chlorine**: 0.2 â€“ 0.5 mg/L  
- **Redox**: > 650 mV (indicates disinfected water)  
- **Conductivity**: < 500 ÂµS/cm (0.0005 S/cm)

 *Source:* WHO

#### 2. Convert data per seconds to per day

```python
daily_Tp_values=wq.groupby(wq['Date'])['Tp'].mean()
daily_pH_values=wq.groupby(wq['Date'])['pH'].mean()
daily_Cl_values=wq.groupby(wq['Date'])['Cl'].mean()
daily_Cl_2_values=wq.groupby(wq['Date'])['Cl_2'].mean()
daily_Redox_values=wq.groupby(wq['Date'])['Redox'].mean()
daily_Leit_values=wq.groupby(wq['Date'])['Leit'].mean()
daily_Trueb_values=wq.groupby(wq['Date'])['Trueb'].mean()
daily_Fm_values=wq.groupby(wq['Date'])['Fm'].mean()
daily_Fm_2_values=wq.groupby(wq['Date'])['Fm_2'].mean()
```

```python
Daily_values=pd.DataFrame({'Tp':daily_Tp_values,'pH':daily_pH_values,'Cl':daily_Cl_values,'Cl_2':daily_Cl_2_values,'Redox':daily_Redox_values,'Leit':daily_Leit_values,'Trueb':daily_Trueb_values,'Fm':daily_Fm_values,'Fm_2':daily_Fm_2_values})
```

```python
dv=Daily_values
```

```python
dv=dv.reset_index()
```

```python
dv
```

#### 3. Visualization of the change of parameters over time

```python
dv.set_index('Date', inplace=True)
```

```python
dv.plot(subplots=True, figsize=(10, 12))
plt.show()
```

**Graphs Explanation:**

1. **Tp:** Temperature is controlled by the company
2. **pH:** The pH graph indicates a stable range of 7.0-8.0, with minor fluctuations due to being affected by temperature. It is safe to drink within the safe range, with no scaling risk.
3. **Cl:** The graph shows that residual chlorine, in the range of 0.10 to 0.40 mg/L, is safe to drink. The sudden drops are due to raise in the flow rate.
4. **Redox:** The graph shows a high redox potential, indicating that the product is safe to drink. It depends on its location if its near the sensor of the first chlore or the second one but it appears that  its affected by the second one, when drops it also drops.
5. **Leit:** The graph displays stable conductivity with occasional small peaks due to external factors, suggesting no pollution, and indicates that it is safe to drink.
6. **Trueb:** The graph shows mostly low turbidity (<1 NTU) with occasional spikes due to drops in chlore, indicating good filtration and safe drinking.
7. **Fm:** Flow rate of the water is controlled by the company.

#### 4. Result of Change

All changes of parameters in the three months are within the safety standards for a drinking water

```python
pH_m=dv['pH'].mean() 
cl_m=dv['Cl'].mean()
redox_m=dv['Redox'].mean() 
leit_m=dv['Leit'].mean() 
trueb_m=dv['Trueb'].mean() 
cl2_m=dv['Cl_2'].mean()
```
