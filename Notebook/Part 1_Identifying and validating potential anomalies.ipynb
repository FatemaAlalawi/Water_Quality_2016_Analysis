# Identifying and validating potential anomalies

```python
e=wq[wq['EVENT']] #extracting anomalies readings
e
```

```python
pH_m=e['pH'].mean()  # mean calculation for water quality parameters for event dataset 
cl_m=e['Cl'].mean()
redox_m=e['Redox'].mean() 
leit_m=e['Leit'].mean() 
trueb_m=e['Trueb'].mean() 
cl2_m=e['Cl_2'].mean()
```

```python
cl_md=e['Cl'].median() # median calculation for water quality parameters for event dataset 
cl2_md=e['Cl_2'].median()
pH_md=e['pH'].median()
redox_md=e['Redox'].median()
trueb_md=e['Trueb'].median()
leit_md=e['Leit'].median()
```

```python
# creating a dictionary of mean and meadian for water quality parameters for event dataset 
mean_data={'Cl':[cl_m,cl_md],'Cl_2':[cl2_m,cl2_md],'pH':[pH_m,pH_md],'Redox':[redox_m,redox_md],'Leit':[leit_m,leit_md],'Trueb':[trueb_m,trueb_md]}
```

```python
md=pd.DataFrame(mean_data) #framing the dictionary
```

```python
Mean_Median_Data_event=md.rename(index={0:'Event Mean',1:'Event Median'}) # renaming the index
Mean_Median_Data_event 
```

## safety and recommended drinking water standards by World Health Organization (WHO)
|Parameter|	Recommended Range|
|---------|------------------|
|pH|6.5 – 8.5|
|Turbidity|	< 5 NTU (ideally < 1 NTU)|
|Chlorine | 0.2 – 0.5 mg/L|
|Redox	|> 650 mV (indicates disinfected water)|
|Conductivity|	< 500 µS/cm (0.0005 S/cm)|


```python
# check if the mean and median values for  water quality parameters for event dataset are within the safety and recommended water drinking standards or not.
standard_check_for_main_parameters_event_data=pd.DataFrame({'Cl':(Mean_Median_Data_event['Cl'] >= 0.2) | (Mean_Median_Data_event['Cl'] <= 0.5),'Cl_2': (Mean_Median_Data_event['Cl_2'] >= 0.2) | (Mean_Median_Data_event['Cl_2'] <= 0.5),'pH': (Mean_Median_Data_event['pH'] >= 6.5) | (Mean_Median_Data_event['pH'] <= 8.5),'Redox':(Mean_Median_Data_event['Redox'] >650),'Leit':(Mean_Median_Data_event['Leit'] <500), 'Trueb': (Mean_Median_Data_event['Trueb'] < 1)})
ch_e=standard_check_for_main_parameters_event_data 
ch_e
```

By comparing the event mean and median values with the standards of drinkability of water, we can say that even when there are changes in Cl, Cl_2, pH, Redox, Leit, and Trueb, water still drinkable and perfectly meets the standards.

```python
pH2_m=nd['pH'].mean() # mean calculation for water quality parameters for normal dataset
cl2_m=nd['Cl'].mean()
redox2_m=nd['Redox'].mean() 
leit2_m=nd['Leit'].mean() 
trueb2_m=nd['Trueb'].mean() 
cl22_m=nd['Cl_2'].mean()
Fm_2=nd['Fm'].mean()
Fm2_2=nd['Fm_2'].mean()
```

```python
pH2_md=nd['pH'].median() # mediancalculation for water quality parameters for normal dataset
cl2_md=nd['Cl'].median() 
redox2_md=nd['Redox'].median() 
leit2_md=nd['Leit'].median() 
trueb2_md=nd['Trueb'].median() 
cl22_md=nd['Cl_2'].median()
```

```python
# creating a dictionary of mean and meadian for water quality parameters for normal dataset 
mean_data_2={'Cl':[cl2_m,cl2_md],'Cl_2':[cl22_m,cl22_md],'pH':[pH2_m,pH2_md],'Redox':[redox2_m,redox2_md],'Leit':[leit2_m,leit2_md],'Trueb':[trueb2_m,trueb2_md]}
```

```python
md2=pd.DataFrame(mean_data_2) #framing the dictionary
```

```python
Mean_Median_Data_normal=md2.rename(index={0:'Normal Mean',1:'Normal Median'}) #renaming index
Mean_Median_Data_normal
```

```python
Mean_Median_Data_event
```

```python
combin=pd.concat([Mean_Median_Data_normal, Mean_Median_Data_event]) #cmbining mean and median data for event and normal dataset
combin
```

```python
#plotting mean and median values of 'Cl','Cl_2', and 'Trueb' for normal and event dataset
combin[['Cl','Cl_2','Trueb']].plot(kind='bar', figsize=(12,7))
plt.title('Comparison of Normal Mean and Event Mean for Water Quality Parameters')
plt.ylabel('Value')
plt.xlabel('Parameter');
```

```python
#plotting mean and median values of 'Redox','Leit', and 'pH'  for normal and event dataset

combin[['Redox','Leit','pH']].plot(kind='bar', figsize=(12,7))
plt.title('Comparison of Normal Mean and Event Mean for Water Quality Parameters')
plt.ylabel('Value')
plt.xlabel('Parameter');
```

By comparing normal data and event data we can conclude that both event and normal data are almost the same with a slight drop in event mean and meadian values, but even though all still within the standarts.

the "event" data generally shows slightly different mean values for several parameters compared to the "normal" data, particularly for pH and Redox, while other parameters like Leit and Trueb remain relatively similar.
pH, redoxidity, cl and cl2 mean values are lower in the event data, while conductivity and Trub mean values are similar which means were not significantly affected by the event.

```python

# we need to convert dataset per seconds to be per day for better understanding and visualisation

daily_Tp_values=wq.groupby(wq['date'])['Tp'].mean()
daily_pH_values=wq.groupby(wq['date'])['pH'].mean()
daily_Cl_values=wq.groupby(wq['date'])['Cl'].mean()
daily_Cl_2_values=wq.groupby(wq['date'])['Cl_2'].mean()
daily_Redox_values=wq.groupby(wq['date'])['Redox'].mean()
daily_Leit_values=wq.groupby(wq['date'])['Leit'].mean()
daily_Trueb_values=wq.groupby(wq['date'])['Trueb'].mean()
daily_Fm_values=wq.groupby(wq['date'])['Fm'].mean()
daily_Fm_2_values=wq.groupby(['date'])['Fm_2'].mean()

```

```python
Daily_values=pd.DataFrame({'Tp':daily_Tp_values,'pH':daily_pH_values,'Cl':daily_Cl_values,'Cl_2':daily_Cl_2_values,'Redox':daily_Redox_values,'Leit':daily_Leit_values,'Trueb':daily_Trueb_values,'Fm':daily_Fm_values,'Fm_2':daily_Fm_2_values})
```
